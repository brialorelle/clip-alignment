```python
good_text_clip_train = "A photo of a dog"
good_image = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/dog.png"
good_image2 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/cat_1.jpg"
good_image3 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/cat_2_carton.jpg"
good_image4 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/cat_3.jpg"
good_image5 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/cat_4.jpg"

good_text_cat = "A photo of a cat"

random_text = "6w!Yc4*zGY3ejfPT"
bad_image_from_babyview_1 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/bad1.jpg"
bad_image_from_babyview_2 = "/home/yinzi/workspace/clip-alignment/babyview_exps/simulate/bad2.jpg"
```

```python
from clip_client import Client
# Initialize the CLIP client
client = Client("grpc://0.0.0.0:51000")
```

```python
from matplotlib import pyplot as plt
from PIL import Image
import numpy as np
```

```python
def show_alignment(image, text):
    text_embedding = client.encode([text])[0]
    text_embedding /= np.linalg.norm(text_embedding)  # Normalization

    image_embedding = client.encode([image])[0]
    image_embedding /= np.linalg.norm(image_embedding)  # Normalization

    similarity = np.dot(text_embedding, image_embedding)
    plt.imshow(Image.open(image))
    plt.title(f"Text: {text}\nSimilarity: {similarity.item()}")
    plt.axis('off')
    plt.show()
```

```python
show_alignment(good_image, good_text_cat)
show_alignment(good_image, "This is a cat")
show_alignment(good_image, "image of animal")
show_alignment(good_image, "image of human")
```

# Aligment of good_text and good_image

```python
show_alignment(good_image, good_text_clip_train)
show_alignment(good_image2, good_text_cat)
show_alignment(good_image3, good_text_cat)
show_alignment(good_image4, good_text_cat)
show_alignment(good_image5, good_text_cat)
```
